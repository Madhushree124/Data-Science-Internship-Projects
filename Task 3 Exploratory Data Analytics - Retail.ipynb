```python
fig,axs = plt.subplots(ncols=2,nrows=2,figsize=(10,10))
sns.distplot(df['Sales'],color='purple',ax= axs[0][0])
sns.distplot(df['Quantity'],color='blue',ax = axs[0][1])
sns.distplot(df['Profit'],color='green',ax= axs[1][0])
sns.distplot(df['Discount'],color='red',ax= axs[1][1])
axs[0][0].set_title('Sales Distribution', fontsize = 20)
axs[0][1].set_title('Quantity Distribution', fontsize = 20)
axs[1][0].set_title('Profit Distribution', fontsize = 20)
axs[1][1].set_title('Discount Distribution', fontsize = 20)
plt.show()
```
